```python
import cv2
import os 

path = 'raw_images'
list_files = os.listdir(path)

file_index = 0
cnt = list_files.count
for file_name in list_files:
    file_path = path + '/' + file_name
    mat = cv2.imread(file_path)
    width = mat.shape[0] / 2
    height = mat.shape[1] / 2
    mat = cv2.resize(mat, (int(height), int(width)))
    cv2.imwrite(file_path, mat)
    print(file_index ,'/' , cnt)
    file_index += 1

print("Finished !")
```
